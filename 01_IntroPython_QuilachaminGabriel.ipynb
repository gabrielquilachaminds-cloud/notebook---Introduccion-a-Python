```python
import pandas as pd
# Creación de un DataFrame manual (Simulando una base de datos)
datos_empresa = {
'Producto': ['Laptop', 'Mouse', 'Monitor', 'Teclado'],
'Precio': [1200, 25, 200, 45],
'Stock': [15, 100, 20, 50]
}
df = pd.DataFrame(datos_empresa)
# Visualización de los primeros registros
print("Vista previa del Sistema de Inventario:")
print(df.head())
# Operación básica: Calcular el valor total del inventario
total_inventario = (df['Precio'] * df['Stock']).sum()
print(f"\nValor total en activos de hardware: ${total_inventario}")
```

```python
import pandas as pd

df = pd.read_csv("/titanic.csv")
```

```python
df.describe()
```

```python
porcentaje = df ["Survived"].mean()*100
print (f"El porcentaje de supervivencia es: {porcentaje:.2f} %")
```

```python
porcentaje_clases = df['Pclass'].value_counts(normalize=True).sort_index() * 100

for clase, valor in porcentaje_clases.items():
    print(f"Clase {clase}: {valor:.2f}%")
```

```python
conteo_genero = df['Sex'].value_counts()

for genero, cantidad in conteo_genero.items():
  if genero == "male":
    nombre = "Masculinos"
  else:
    nombre = "Femeninos"

  print(f"Cantidad de pasajeros {nombre}: {cantidad}")
```

```python
# Conclusiones:
## 1. Porcentaje de supervivencia:
# *El porcentaje de supervivencia fue de 38.38%, lo que muestra que menos de la mitad de las personas sobrevivieron en el accidente.
# *Notamos que en la parte de supervivencia del csv estaba una notación de 0 y 1, lo que nos permitió identificar las peronas que sobrevivieron (1) y las que no (0).

## 2. Porcentaje de pasajeros según la clase:
# * La mayoría de los pasajeros viajaban en tercera clase, lo que en porcentaje representa el 55.11%
# * Notamos que este dato no estaba en la columna de Parch, como se mencionaba en la guia, se encontraba en la sección de Pclass.

## 3. Conteo de géneto:
# * La mayoría de personas que se encontraban viajando en el titanic fueron de género masculino, con un total de 577; por parte del género femenino viajaban 314.
```
